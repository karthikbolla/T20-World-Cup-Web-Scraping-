```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv

url = 'https://www.t20worldcup.com/fixtures'
p = requests.get(url)
soup = BeautifulSoup(p.content,'html.parser')
content = soup.find_all()
print(p)

content = soup.find_all('time', class_="match-block__date-local")
print(content[0].text)

content1 = soup.find_all('div', class_="match-block__teams-wrapper match-block__teams-wrapper--desktop")

frontheader = ["date","match name"]
itemfull = []

dates = []
for item in content:
  dates.append(item.text[3:].strip())

print(dates)

for item in content1:
  items = []
  content2 = item.find_all('span', class_="match-block__team-name")
  matchname = content2[0].text+" Vs "+content2[1].text
  itemfull.append(matchname)
print(itemfull)  

fulldata = []
for i in zip(dates,itemfull):
  fulldata.append(i)

print(fulldata)

pd.DataFrame(fulldata).to_csv("/content/sample_data/t20 worldcup.csv",header=frontheader)
```
